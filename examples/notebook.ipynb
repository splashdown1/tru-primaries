```python
from tru import TRU

tru = TRU()
stats = tru.stats()
print(f"Primaries: {stats.total_cached}")
print(f"SEC filings: {stats.sec_filings}")
print(f"arXiv papers: {stats.arxiv_papers}")
```

```python
# Search narraws from tru import TRU
tru = TRU()

results = tru.search("attention mechanism", source_type="arxiv", limit=10)
for r in results:
    print(f"{r.title[:60]}...")
```

```python
# Verify claims
claims = [
    "NVIDIA has achieved AGI",
    "The Temple Institute monitors red heifers"
]

for claim in claims:
    v = tru.verify(claim)
    print(f"{v.verdict}: {claim}")
```
